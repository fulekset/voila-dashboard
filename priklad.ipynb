```python
import matplotlib.pyplot as plt
import numpy as np

# Generovanie náhodných údajov
data = np.random.randn(1000)

# Vytvorenie histogramu
plt.hist(data, bins=30, edgecolor='black')

# Pridanie názvov osi a nadpisu
plt.xlabel('Hodnoty')
plt.ylabel('Frekvencia')
plt.title('Histogram')

# Zobrazenie grafu
plt.show()
```
